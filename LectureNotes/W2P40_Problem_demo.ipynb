```python
import numpy as np
from scipy.stats import binom
```

```python
n, p = 10, 0.25
```

```python
N_TRIALS = int(1e6)

total = 0
n_tests = 2
for i in range(N_TRIALS):
     total += max(binom.rvs(n, p, size=n_tests))
avg = total / N_TRIALS
print(avg)
```

```python
rvs = np.array(binom.rvs(n, p, size=n_tests * N_TRIALS)).reshape((n_tests, -1))
rvs.max(axis=0).mean()
```

```python
np.sum(1 - binom.cdf(np.arange(0, 11), n, p) ** n_tests) / 10
```
