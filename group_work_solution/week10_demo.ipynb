```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rcParams.update({'font.size': 16})
fig, ax = plt.subplots(figsize=(8,8))
circle1 = plt.Circle((0, 0), np.sqrt(2), fill=False, edgecolor='blue', linewidth=2, linestyle='--')
ax.plot([0, 1], [0, 1], linewidth=2, color='blue')
ax.plot([0, 1], [0, 0], linewidth=2, color='red')
ax.plot([1, 1], [0, 1], linewidth=2, color='red')
ax.plot([-1, -1], [-1, 1], linewidth=2, color='red', linestyle='--')
ax.plot([-1, 1], [1, 1], linewidth=2, color='red', linestyle='--')
ax.plot([-1, 1], [-1, -1], linewidth=2, color='red', linestyle='--')
ax.plot([1, 1], [-1, 1], linewidth=2, color='red', linestyle='--')
ax.add_patch(circle1)
ax.set_xlim((-1.5, 1.5))
ax.set_ylim((-1.5, 1.5))
plt.grid()
plt.tight_layout()
plt.savefig('distance_plot.pdf', dpi=300)
```
