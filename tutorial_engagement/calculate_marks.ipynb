```python
import os

import pandas as pd
import numpy as np
```

```python
path_file = 'COMPSCI762_2022S1_TutorialEngagement - Sheet1.csv'
with open(path_file, 'r') as f:
  lines = f.readlines()
  print(lines[1:])
```

```python
def update_scores(name: str, scores: set) -> None:
  if name in scores:
    if scores[name] < 2:
      scores[name] += 1
  else:
    scores[name] = 1
```

```python
scores = {}
for line in lines:
  names = line.split(',')
  names = names[2:-1]
  names = [i for i in names if i]
  for name in names:
    update_scores(name, scores)

print(scores)
```

```python
df = pd.DataFrame(list(scores.items()), columns=['UPI', 'Score'])
df['UPI'] = df['UPI'].str.lower()
df = df.sort_values(by=['UPI'])
df = df.set_index('UPI')
df
```

```python
df.to_csv('student_scores.csv')
```
