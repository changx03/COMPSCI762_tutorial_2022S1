```python
import json
import pandas as pd
```

```python
path_file = 'COMPSCI762_2022S1_TutorialEngagement - Sheet1.csv'
with open(path_file, 'r') as f:
  lines = f.readlines()
  print(lines[1:])
```

```python
def update_scores(name: str, scores: set) -> None:
  if name in scores:
    if scores[name] < 2:
      scores[name] += 1
  else:
    scores[name] = 1
```

```python
scores = {}
for line in lines:
  names = line.split(',')
  names = names[2:-1]
  names = [i for i in names if i]
  for name in names:
    update_scores(name, scores)

print(scores)
```

```python
df = pd.DataFrame(list(scores.items()), columns=['UPI', 'Score'])
df['UPI'] = df['UPI'].str.lower()
df = df.set_index('UPI')
df
```

```python
path_jsons = [f'package{i}.json' for i in range(1, 4)]
print(path_jsons)

people = []
for p in path_jsons:
  with open(p) as f:
    people = people + json.load(f)

print(len(people))
print(people[0])
```

```python
people_dict = {}
for person in people:
  upi = person['login_id']
  name = person['sortable_name']
  people_dict[upi] = name
```

```python
names = [people_dict[i] for i in df.index]
df['name'] = names
df = df.sort_values(by=['name'])
df.head()
```

```python
df.to_csv('student_scores.csv')
```
